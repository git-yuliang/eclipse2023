### **Notebook Overview**

These Jupyter Notebook documents the processing and analysis of observational data collected during the 2023 total solar eclipse. The primary objectives of this Notebook include:

1. Preprocessing of raw white-light corona imaging data (dark field subtraction, flat field correction, and temporal registration).
2. Application of wavelet analysis to detect high-frequency quasi-periodic oscillations (HFQPOs).
3. Visualizations of the solar corona and statistical results for Sun-as-a-star analysis.

### **Notebook figure - plot of the wavelet analysis results**

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Update plotting parameters
plt.rcParams.update({
    'font.size': 8,
    'axes.labelsize': 8,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'font.family': 'serif',
    'font.serif': ['Times New Roman'],  # Using Times New Roman font
    'mathtext.fontset': 'stix',
})

# Load the images
img_r11 = mpimg.imread('./output/intensity_r11.png')
img_bkg = mpimg.imread('./output/intensity_bkg.png')
img_fov = mpimg.imread('./output/intensity_fov.png')

# Set up the figure for stacking the images vertically
fig, axes = plt.subplots(3, 1, figsize=(7.09, 6.0), dpi=600)

# Plot each image in its respective subplot
axes[0].imshow(img_r11)
axes[0].axis('off')  # Turn off axis
axes[0].set_xlabel('Frame no.', fontsize=8)  # Set label with proper case
axes[0].set_ylabel('Intensity / DN', fontsize=8)

axes[1].imshow(img_bkg)
axes[1].axis('off')  # Turn off axis
axes[1].set_xlabel('Frame no.', fontsize=8)  # Set label with proper case
axes[1].set_ylabel('Background intensity', fontsize=8)

axes[2].imshow(img_fov)
axes[2].axis('off')  # Turn off axis
axes[2].set_xlabel('Frame no.', fontsize=8)  # Set label with proper case
axes[2].set_ylabel('FOV intensity', fontsize=8)

# Adjust layout
plt.tight_layout()

# Save the combined image
pathfigs = './output/'
plt.savefig(pathfigs + 'detrend_normalized.png', bbox_inches='tight')

# Show the combined image
plt.show()
```
